# import packages (numpy, matplotlib) for computing function & plotting graph

```python
import numpy as np
import matplotlib.pyplot as plt
```

# define 2-dimensional differentiable function: $f(x) = \sin^2(x) \cdot x$

```python
def my2DimFunc(x):
    f= np.sin(x) * np.sin(x) * x
    return f
```

# define the derivation of given function: $f'(x) = \sin^2(x) + \sin(2\cdot x)\cdot x$

```python
def DerivativeFunc(x):
    f= np.sin(x) * np.sin(x) + np.sin(2*x) * x
    return f
```

# define the domain: $x = [-10 : 0.1 : 10]$

```python
x = np.arange(-10,10,0.1)
```

# pick 3 points in the domain and compute the function value:
# $p1=(-7, f(-7)), p2=(-5,  f(-5)), p3=(3,  f(3))$
# $p'1=(-7, f'(-7)), p'2=(-5, f'(-5)), p'3=(3, f'(3))$

```python
p = np.array([-7, -5, 3])
fv = my2DimFunc(p)
Dfv = DerivativeFunc(p)
```

# insert the domain to defined function

```python
myf = my2DimFunc(x)
myDf = DerivativeFunc(x)
```

# plot the graph of defined function, derivative function and given points

```python
plt.figure(1)
plt.plot(x, myf, 'b', label="function")
plt.plot(x, myDf, 'g', label="derivative")
plt.plot(p, fv, 'ro')
plt.plot(p, Dfv, 'yo')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

# define Taylor approximation of given function: $\hat{f}(x) = f(z) + \nabla f(z)^T (x-z)$

```python
def myTaylor(x, p):
    f = my2DimFunc(p)+DerivativeFunc(p)*(x-p)
    return f
```

# compute Taylor approximation of given points

```python
Tf1 = myTaylor(x, p[0])
Tf2 = myTaylor(x, p[1])
Tf3 = myTaylor(x, p[2])
```

# plot Taylor approximation of given points

```python
plt.figure(2)
plt.plot(x, myf, 'b', label="function")
plt.plot(x, Tf1, 'r', label="taylor of p1")
plt.plot(x, Tf2, 'g', label="taylor of p2")
plt.plot(x, Tf3, 'y', label="taylor of p3")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
axes = plt.gca()
axes.set_xlim([-10, 10])
axes.set_ylim([-10, 10])
```
